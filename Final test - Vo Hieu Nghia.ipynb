```python
# My code
from scipy.stats import norm
#a
rv = norm(loc = 120, scale = 20)
xrnd = rv.rvs(size = 10000)

hist(xrnd, bins = arange(40,200,1.65), alpha = 1)
xlabel('Numbers')
ylabel('Frequency (#)')
grid()
print(xrnd)

#b
count = 0
count2 = 0
for i in range (len(xrnd)):
    if xrnd[i]>160:
        count = count+1
print('Probability > 160 is :',count/10000)

#c
for i in range (len(xrnd)):
    if xrnd[i]>=80 and xrnd[i]<=140 :
        count2 = count2+1
print('Probability between 80 and 140 is', count2/10000)
```
