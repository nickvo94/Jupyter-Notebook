Create an `x`-array having values from 0 to 6 (stepsize = 0.1) and use that to calculate the values for $y = (x-3)^2 - 10$. Make a plot of that function and annotate the graphics with labels and title.

```python
x = np.arange(0,6.1,0.1)
xsize = x.size
y = np.zeros(xsize)
print(x)
for i in range (0,xsize):
    valx = x[i]
    valy = ((valx-3)**2)-10
    y[i] = valy
print(y)    
fig = plt.figure()
ax = fig.add_subplot(111) 
ax.plot(x, y, color='lightblue', linewidth=3)
ax.set_xlim(0, 6)

ax.set(title='Fuction graph', ylabel='Y-Axis', xlabel='X-Axis')

    
plt.savefig('1.png')
plt.show()
```

Use the same data to create the following 2D graphs:

a) a step-like graph   
b) a bar graph   
c) a scatterogram   
d) an area filled between the function and the x-axis

```python
x = np.arange(0,6.1,0.1)
xsize = x.size
y = np.zeros(xsize)
for i in range (0,xsize):
    valx = x[i]
    valy = ((valx-3)**2)-10
    y[i] = valy
    
fig = plt.figure()
ax = fig.add_subplot(111)
ax.step(x,y)
ax.set_xlim(0, 6)
plt.savefig('2a.png')
plt.show()

fig2 = plt.figure()
ax = fig2.add_subplot(111)
ax.bar(x,y, facecolor='#9999ff', edgecolor='white')
ax.set_xlim(0, 6)
plt.savefig('2b.png')
plt.show()

fig3 = plt.figure()
ax = fig3.add_subplot(111)
ax.scatter(x,y)
ax.set_xlim(0, 6)
plt.savefig('2c.png')
plt.show()

fig4 = plt.figure()
ax = fig4.add_subplot(111)
ax.fill_between(x,y)
ax.set_xlim(0, 6)
plt.savefig('2d.png')
plt.show()
```

Plot in the same graph the following approximations of log(x):

$f_1(x) = (x-1)$    
$f_2(x) = (x-1) - \frac{1}{2}(x-1)^2$   
$f_3(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3$   
$f_4(x) = (x-1) - \frac{1}{2}(x-1)^2 + \frac{1}{3}(x-1)^3 - \frac{1}{4}(x-1)^4$   

Use the range $x \in [0, 2]$ with suitable stepsize. Add legends to the graphs to explain what the graphs are.

```python
x = np.arange(0,2,0.05)
xsize = x.size
y1 = np.zeros(xsize)
y2 = np.zeros(xsize)
y3 = np.zeros(xsize)
y4 = np.zeros(xsize)

for i in range (0,xsize):
    valx = x[i]
    valy1 = valx-1
    y1[i] = valy1
    
    valy2= valy1 - (1/2)*((valy1)**2)
    y2[i] = valy2
    
    valy3 = valy2 + (1/3)*((valy1)**3)
    y3[i] = valy3
    
    valy4 = valy3 - (1/4)*((valy1)**4)
    y4[i] = valy4
    
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, color='r', linewidth=3)
ax.plot(x, y2, color='g', linewidth=3)
ax.plot(x, y3, color='b', linewidth=3)
ax.plot(x, y4, color='black', linewidth=3)
ax.set_xlim(0, 2.05)

ax.set(title='Fuction log(x)', ylabel='Y-Axis', xlabel='X-Axis')
    
plt.savefig('4.png')
plt.show()
```
